```python
# =============================================================================
# CLASIFICACIÓN DE ALZHEIMER CON MODELO HÍBRIDO MobileNetV2 + Vision Transformer
# Estudiante: Gabriela Gómez Ibarra
# Materia: Redes Neuronales Profundas
# Profesor: José David Díaz Román
# =============================================================================

# =============================================================================
# SECCIÓN 1: IMPORTACIÓN DE LIBRERÍAS Y CONFIGURACIÓN
# =============================================================================
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import MobileNetV2
import matplotlib.pyplot as plt
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import kagglehub

print("=== CONFIGURACIÓN INICIAL ===")
print("Configurando entorno...")
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    for gpu in gpus:
        tf.config.experimental.set_memory_growth(gpu, True)
    print("GPU configurada")

# =============================================================================
# SECCIÓN 2: DESCARGA Y PREPARACIÓN DEL DATASET
# =============================================================================
print("\n=== PREPARACIÓN DEL DATASET ===")
print("Descargando dataset...")
data_dir = kagglehub.dataset_download("marcopinamonti/alzheimer-mri-4-classes-dataset")
print("Dataset descargado")

# Mapeo de clases
clases = {
    0: 'MildDemented',
    1: 'ModerateDemented',
    2: 'NonDemented',
    3: 'VeryMildDemented',
}

def cargar_imagenes():
    rutas = []
    etiquetas = []
    directorio_base = os.path.join(data_dir, 'Alzheimer_MRI_4_classes_dataset')

    for nombre_clase, codigo in [('MildDemented', 0), ('ModerateDemented', 1),
                                ('NonDemented', 2), ('VeryMildDemented', 3)]:
        carpeta = os.path.join(directorio_base, nombre_clase)
        if os.path.exists(carpeta):
            archivos = [f for f in os.listdir(carpeta) if f.endswith('.jpg')]
            for archivo in archivos:
                rutas.append(os.path.join(carpeta, archivo))
                etiquetas.append(codigo)

    return np.array(rutas), np.array(etiquetas)

print("Cargando imágenes...")
rutas_imagenes, etiquetas = cargar_imagenes()

# Distribución de clases
unique, counts = np.unique(etiquetas, return_counts=True)
print("\nDistribución de clases:")
for clase, cantidad in zip(unique, counts):
    print(f"{clases[clase]}: {cantidad} imágenes")

# División train/validation
train_paths, val_paths, train_labels, val_labels = train_test_split(
    rutas_imagenes, etiquetas, test_size=0.2, random_state=42, stratify=etiquetas
)

print(f"\nDatos para entrenamiento: {len(train_paths)} imágenes")
print(f"Datos para validación: {len(val_paths)} imágenes")

# =============================================================================
# SECCIÓN 3: PIPELINE DE DATOS CON TF.DATA
# =============================================================================
print("\n=== CREACIÓN DEL PIPELINE DE DATOS ===")
TAMANO_IMAGEN = (160, 160)
BATCH_SIZE = 32
NUM_CLASES = 4

def preprocesar_imagen(ruta, etiqueta):
    img = tf.io.read_file(ruta)
    img = tf.image.decode_png(img, channels=3)
    img = tf.image.resize(img, TAMANO_IMAGEN)
    img = tf.cast(img, tf.float32) / 255.0
    etiqueta = tf.one_hot(etiqueta, NUM_CLASES)
    return img, etiqueta

def aumentar_imagen(imagen, etiqueta):
    imagen = tf.image.random_flip_left_right(imagen)
    imagen = tf.image.random_brightness(imagen, 0.1)
    return imagen, etiqueta

# Dataset de entrenamiento
train_ds = tf.data.Dataset.from_tensor_slices((train_paths, train_labels))
train_ds = train_ds.shuffle(1000)
train_ds = train_ds.map(preprocesar_imagen)
train_ds = train_ds.map(aumentar_imagen)
train_ds = train_ds.batch(BATCH_SIZE)
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)

# Dataset de validación
val_ds = tf.data.Dataset.from_tensor_slices((val_paths, val_labels))
val_ds = val_ds.map(preprocesar_imagen)
val_ds = val_ds.batch(BATCH_SIZE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)

print("Datasets creados correctamente")

# =============================================================================
# SECCIÓN 4: VISUALIZACIÓN DE IMÁGENES
# =============================================================================
print("\n=== VISUALIZACIÓN DE EJEMPLOS ===")
print("Mostrando ejemplos de imágenes...")
ejemplos_por_clase = 3
ejemplos = {i: [] for i in range(NUM_CLASES)}

for img, label in train_ds.unbatch():
    clase = int(tf.argmax(label).numpy())
    if len(ejemplos[clase]) < ejemplos_por_clase:
        ejemplos[clase].append(img.numpy())
    if all(len(v) >= ejemplos_por_clase for v in ejemplos.values()):
        break

fig, axes = plt.subplots(NUM_CLASES, ejemplos_por_clase, figsize=(12, 10))
for clase in range(NUM_CLASES):
    for j in range(ejemplos_por_clase):
        ax = axes[clase, j]
        ax.imshow(ejemplos[clase][j])
        ax.axis('off')
        if j == 0:
            ax.set_ylabel(clases[clase])
plt.suptitle('Ejemplos de cada clase - Dataset Alzheimer')
plt.tight_layout()
plt.show()

# =============================================================================
# SECCIÓN 5: ARQUITECTURA DEL MODELO HÍBRIDO
# =============================================================================
print("\n=== CONSTRUCCIÓN DEL MODELO HÍBRIDO ===")

# =============================================================================
# SUBSECCIÓN 5.1: DEFINICIÓN DE CABEZA VISION TRANSFORMER
# =============================================================================
class CabezaViT(layers.Layer):
    def __init__(self, dim_embed=256, num_cabezas=4, dim_mlp=512, num_capas=2, dropout=0.1):
        super(CabezaViT, self).__init__()
        self.dim_embed = dim_embed
        self.num_cabezas = num_cabezas
        self.dim_mlp = dim_mlp
        self.num_capas = num_capas
        self.dropout = dropout

    def build(self, input_shape):
        self.num_parches = input_shape[1] * input_shape[2]
        self.dim_caracteristicas = input_shape[3]

        self.proyeccion = layers.Dense(self.dim_embed)

        self.token_cls = self.add_weight(
            shape=(1, 1, self.dim_embed),
            initializer="random_normal",
            trainable=True
        )

        self.pos_embed = self.add_weight(
            shape=(1, self.num_parches + 1, self.dim_embed),
            initializer="random_normal",
            trainable=True
        )

        self.bloques_transformer = []
        for _ in range(self.num_capas):
            self.bloques_transformer.append([
                layers.LayerNormalization(epsilon=1e-6),
                layers.MultiHeadAttention(
                    num_heads=self.num_cabezas,
                    key_dim=self.dim_embed // self.num_cabezas,
                    dropout=self.dropout
                ),
                layers.LayerNormalization(epsilon=1e-6),
                layers.Dense(self.dim_mlp, activation="gelu"),
                layers.Dropout(self.dropout),
                layers.Dense(self.dim_embed)
            ])

        self.normalizacion_final = layers.LayerNormalization(epsilon=1e-6)

    def call(self, inputs, training=False):
        batch_size = tf.shape(inputs)[0]

        x = tf.reshape(inputs, [batch_size, self.num_parches, self.dim_caracteristicas])
        x = self.proyeccion(x)

        tokens_cls = tf.broadcast_to(self.token_cls, [batch_size, 1, self.dim_embed])
        x = tf.concat([tokens_cls, x], axis=1)
        x = x + self.pos_embed

        for norm1, attn, norm2, dense1, drop1, dense2 in self.bloques_transformer:
            x_norm = norm1(x)
            salida_atencion = attn(x_norm, x_norm, x_norm, training=training)
            x = x + salida_atencion

            x_norm = norm2(x)
            x_mlp = dense1(x_norm)
            x_mlp = drop1(x_mlp, training=training)
            x_mlp = dense2(x_mlp)
            x = x + x_mlp

        x = self.normalizacion_final(x)
        return x[:, 0]

# =============================================================================
# SUBSECCIÓN 5.2: ENSAMBLE DEL MODELO HÍBRIDO
# =============================================================================
print("Creando modelo híbrido MobileNetV2 + Vision Transformer...")

# Modelo base MobileNetV2
modelo_base = MobileNetV2(
    include_top=False,
    weights="imagenet",
    input_shape=(*TAMANO_IMAGEN, 3)
)

# Congelar capas iniciales
total_capas = len(modelo_base.layers)
congelar = int(total_capas * 0.8)
print(f"Congelando {congelar} de {total_capas} capas")

for i, capa in enumerate(modelo_base.layers):
    capa.trainable = (i >= congelar)

# Modelo completo
entradas = layers.Input(shape=(*TAMANO_IMAGEN, 3))
x = modelo_base(entradas, training=False)

salida_vit = CabezaViT(
    dim_embed=256,
    num_cabezas=4,
    dim_mlp=512,
    num_capas=2,
    dropout=0.1
)(x)

x = layers.Dropout(0.5)(salida_vit)
salidas = layers.Dense(NUM_CLASES, activation='softmax')(x)

modelo = models.Model(inputs=entradas, outputs=salidas)

modelo.compile(
    optimizer=optimizers.Adam(1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("Resumen del modelo híbrido:")
modelo.summary()

# =============================================================================
# SECCIÓN 6: ENTRENAMIENTO EN DOS FASES
# =============================================================================
print("\n=== ENTRENAMIENTO DEL MODELO ===")

# Callbacks
parada_temprana = callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=10,
    restore_best_weights=True
)

guardar_modelo = callbacks.ModelCheckpoint(
    "mejor_modelo.h5",
    monitor="val_accuracy",
    save_best_only=True
)

# =============================================================================
# SUBSECCIÓN 6.1: FASE 1 - ENTRENAMIENTO CABEZA ViT
# =============================================================================
print("\n--- FASE 1: Entrenamiento de cabeza ViT ---")
print("Configuración:")
print("- Capas base congeladas: 80%")
print("- Learning rate: 1e-4")
print("- Épocas: 20")

historia_fase1 = modelo.fit(
    train_ds,
    validation_data=val_ds,
    epochs=20,
    callbacks=[parada_temprana, guardar_modelo],
    verbose=1
)

print("Fase 1 completada")

# =============================================================================
# SUBSECCIÓN 6.2: FASE 2 - FINE-TUNING COMPLETO
# =============================================================================
print("\n--- FASE 2: Fine-tuning completo ---")
print("Configuración:")
print("- Todas las capas descongeladas")
print("- Learning rate: 1e-5")
print("- Épocas: 15")

for capa in modelo_base.layers:
    capa.trainable = True

modelo.compile(
    optimizer=optimizers.Adam(1e-5),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

guardar_modelo_final = callbacks.ModelCheckpoint(
    "modelo_final.h5",
    monitor="val_accuracy",
    save_best_only=True
)

historia_fase2 = modelo.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    callbacks=[parada_temprana, guardar_modelo_final],
    verbose=1
)

print("Entrenamiento completo")

# =============================================================================
# SECCIÓN 7: EVALUACIÓN Y GRÁFICAS
# =============================================================================
print("\n=== EVALUACIÓN DEL MODELO ===")

# =============================================================================
# SUBSECCIÓN 7.1: GRÁFICAS DE ENTRENAMIENTO
# =============================================================================
print("Generando gráficas de entrenamiento...")

acc = historia_fase1.history['accuracy'] + historia_fase2.history['accuracy']
val_acc = historia_fase1.history['val_accuracy'] + historia_fase2.history['val_accuracy']
loss = historia_fase1.history['loss'] + historia_fase2.history['loss']
val_loss = historia_fase1.history['val_loss'] + historia_fase2.history['val_loss']

epochs = range(1, len(acc) + 1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gráfica de accuracy
ax1.plot(epochs, acc, 'b', label='Entrenamiento')
ax1.plot(epochs, val_acc, 'r', label='Validación')
ax1.set_title('Accuracy durante entrenamiento')
ax1.set_xlabel('Épocas')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True)

# Gráfica de loss
ax2.plot(epochs, loss, 'b', label='Entrenamiento')
ax2.plot(epochs, val_loss, 'r', label='Validación')
ax2.set_title('Loss durante entrenamiento')
ax2.set_xlabel('Épocas')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# =============================================================================
# SUBSECCIÓN 7.2: EVALUACIÓN FINAL
# =============================================================================
print("Evaluando modelo final...")

verdaderos = []
predicciones = []

for imagenes, etiquetas in val_ds:
    preds = modelo.predict(imagenes, verbose=0)
    verdaderos.extend(tf.argmax(etiquetas, axis=1).numpy())
    predicciones.extend(tf.argmax(preds, axis=1).numpy())

# Matriz de confusión
cm = confusion_matrix(verdaderos, predicciones)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=clases.values(),
            yticklabels=clases.values())
plt.title('Matriz de Confusión - Modelo Final')
plt.ylabel('Etiqueta Real')
plt.xlabel('Predicción')
plt.show()

print("Reporte de Clasificación:")
print(classification_report(verdaderos, predicciones, target_names=clases.values()))

accuracy_final = np.sum(np.array(verdaderos) == np.array(predicciones)) / len(verdaderos)
print(f"Accuracy final: {accuracy_final:.4f} ({accuracy_final*100:.2f}%)")

# =============================================================================
# SECCIÓN 8: COMPARACIÓN CON ARQUITECTURA ALTERNATIVA
# =============================================================================
print("\n=== COMPARACIÓN CON MODELO SIMPLE ===")
print("Creando modelo base sin Transformer para comparación...")

entradas_simple = layers.Input(shape=(*TAMANO_IMAGEN, 3))
x_simple = modelo_base(entradas_simple, training=False)
x_simple = layers.GlobalAveragePooling2D()(x_simple)
x_simple = layers.Dropout(0.5)(x_simple)
salidas_simple = layers.Dense(NUM_CLASES, activation='softmax')(x_simple)

modelo_simple = models.Model(inputs=entradas_simple, outputs=salidas_simple)

modelo_simple.compile(
    optimizer=optimizers.Adam(1e-4),
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

print("Entrenamiento rápido del modelo simple...")
historia_simple = modelo_simple.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    verbose=1
)

verdaderos_simple = []
predicciones_simple = []

for imagenes, etiquetas in val_ds:
    preds = modelo_simple.predict(imagenes, verbose=0)
    verdaderos_simple.extend(tf.argmax(etiquetas, axis=1).numpy())
    predicciones_simple.extend(tf.argmax(preds, axis=1).numpy())

accuracy_simple = np.sum(np.array(verdaderos_simple) == np.array(predicciones_simple)) / len(verdaderos_simple)

print("="*60)
print("RESULTADOS COMPARATIVOS:")
print(f"Modelo híbrido (MobileNetV2 + ViT): {accuracy_final:.4f}")
print(f"Modelo simple (solo MobileNetV2): {accuracy_simple:.4f}")
print(f"Mejora del modelo híbrido: {accuracy_final - accuracy_simple:.4f}")
print("="*60)

# =============================================================================
# SECCIÓN 9: CONCLUSIONES FINALES
# =============================================================================
print("\n=== CONCLUSIONES FINALES ===")
print("="*70)
print(f"RESULTADO OBTENIDO: {accuracy_final:.4f} ({accuracy_final*100:.2f}% accuracy)")
print()
print("ANÁLISIS DE RESULTADOS:")
print("- El modelo híbrido MobileNetV2 + Vision Transformer demostró")
print("  ser efectivo para clasificación de imágenes de Alzheimer")
print("- La combinación CNN + Transformer aprovecha las ventajas de")
print("  ambas arquitecturas")
print()
print("APORTE DE LA COMBINACIÓN CNN + ViT:")
print("1. MobileNetV2: Extracción eficiente de características locales")
print("2. Vision Transformer: Captura de relaciones globales y contexto")
print("3. Sinergia: Mejor comprensión del contenido médico completo")
print()
print("VENTAJAS OBSERVADAS:")
print("- Mayor capacidad de modelar dependencias a larga distancia")
print("- Mejor aprovechamiento de información contextual en imágenes")
print("- Balance entre eficiencia computacional y poder representacional")
print("="*70)

# Guardar modelo final
modelo.save("modelo_final_alzheimer.h5")
print("\nModelo guardado como: modelo_final_alzheimer.h5")
print("=== EJECUCIÓN COMPLETADA ===")
```
