```
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import time

# Parámetros
VOCAB_SIZE = 10000  # Top palabras
MAX_LEN = 200       # Longitud fija de secuencia

# Cargar y preprocesar datos
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)
x_train = pad_sequences(x_train, maxlen=MAX_LEN, padding="pre", truncating="pre")
x_test = pad_sequences(x_test, maxlen=MAX_LEN, padding="pre", truncating="pre")
print(f"Train: {x_train.shape}, Test: {x_test.shape}")

# Función para crear y entrenar modelo
def train_model(model_type, units=64):
    model = keras.Sequential([
        keras.layers.Input(shape=(MAX_LEN,), dtype='int32'),
        keras.layers.Embedding(input_dim=VOCAB_SIZE, output_dim=128),
        model_type(units),  # SimpleRNN, LSTM o GRU
        layers.Dense(1, activation="sigmoid")
    ])
    model.compile(optimizer=keras.optimizers.Adam(1e-3),
                  loss="binary_crossentropy",
                  metrics=["accuracy"])

    callbacks = [keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)]
    model.summary()

    start_time = time.time()
    history = model.fit(x_train, y_train,
                        validation_split=0.15,
                        epochs=10,
                        batch_size=64,
                        callbacks=callbacks,
                        verbose=2)
    total_time = time.time() - start_time
    avg_time_per_epoch = total_time / len(history.history['loss'])

    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    print(f"[{model_type.__name__}] Test acc: {test_acc:.4f} | loss: {test_loss:.4f}")
    print(f"Tiempo promedio por época: {avg_time_per_epoch:.2f} segundos")

    return history, avg_time_per_epoch, test_acc, test_loss

# Experimento 1: SimpleRNN (replica el ejemplo original)
print("\nEntrenando SimpleRNN...")
hist_rnn, time_rnn, acc_rnn, loss_rnn = train_model(layers.SimpleRNN)

# Experimento 2: LSTM (solo cambia la línea layers.SimpleRNN(64) por layers.LSTM(64))
print("\nEntrenando LSTM...")
hist_lstm, time_lstm, acc_lstm, loss_lstm = train_model(layers.LSTM)

# Experimento 3: GRU (solo cambia la línea layers.LSTM(64) por layers.GRU(64))
print("\nEntrenando GRU...")
hist_gru, time_gru, acc_gru, loss_gru = train_model(layers.GRU)

# Función para graficar curvas (Accuracy y Loss vs epochs, train/val)
def plot_history(hist, title):
    fig, ax = plt.subplots(1, 2, figsize=(10, 4))
    ax[0].plot(hist.history["loss"], label="train")
    ax[0].plot(hist.history["val_loss"], label="val")
    ax[0].set_title(f"{title} - Loss"); ax[0].set_xlabel("Epoch"); ax[0].set_ylabel("Loss"); ax[0].legend()
    ax[1].plot(hist.history["accuracy"], label="train")
    ax[1].plot(hist.history["val_accuracy"], label="val")
    ax[1].set_title(f"{title} - Accuracy"); ax[1].set_xlabel("Epoch"); ax[1].set_ylabel("Acc"); ax[1].legend()
    plt.tight_layout(); plt.show()

# Genera las curvas para cada red
plot_history(hist_rnn, "SimpleRNN")
plot_history(hist_lstm, "LSTM")
plot_history(hist_gru, "GRU")

# Resumen de desempeño (acc/loss test y tiempo por época)
print("\nResumen de desempeño:")
print(f"SimpleRNN -> Test acc: {acc_rnn:.4f} | Test loss: {loss_rnn:.4f} | Tiempo por época: {time_rnn:.2f} s")
print(f"LSTM -> Test acc: {acc_lstm:.4f} | Test loss: {loss_lstm:.4f} | Tiempo por época: {time_lstm:.2f} s")
print(f"GRU -> Test acc: {acc_gru:.4f} | Test loss: {loss_gru:.4f} | Tiempo por época: {time_gru:.2f} s")
```
